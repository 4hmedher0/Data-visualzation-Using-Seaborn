```python
import pandas as pd 
import numpy as np 
import seaborn as sns
import matplotlib.pyplot as plt 
```

```python
df=pd.read_csv('Iris.csv')
```

```python
d1=df.iloc[0:4]
d2=df.iloc[55:60]
d3=df.iloc[120:125]
df1=pd.concat([d1,d2,d3])
df1.reset_index(inplace=True)
df1.drop(['index','Id'],axis=1,inplace=True)
```

```python
fig=plt.gcf()
fig.set_size_inches(12,8)
ax=sns.boxplot(x='Species',y='SepalLengthCm',data=df1,width=0.5,whis=1,linewidth=3,fliersize=9,color='c',order=['Iris-versicolor','Iris-virginica','Iris-setosa'])
```
