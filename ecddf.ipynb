```python
import pandas as pd 
import numpy as np 
import seaborn as sns
import matplotlib.pyplot as plt 
```

```python
df=sns.load_dataset('mpg').dropna()
```

```python
df.head()
```

```python
fig=plt.gcf()
fig.set_size_inches(12,8)
ax=sns.ecdfplot(x='horsepower',data=df)
```

```python
df.horsepower.min()
```

```python
df.horsepower.max()
```

```python
fig=plt.gcf()
fig.set_size_inches(12,8)
ax=sns.ecdfplot(x='horsepower',data=df)
plt.axhline(0.5,c='r')
```

```python
fig=plt.gcf()
fig.set_size_inches(12,8)
ax=sns.ecdfplot(x='horsepower',data=df)
plt.axvline(100,c='r')
```

```python
fig=plt.gcf()
fig.set_size_inches(12,8)
ax=sns.ecdfplot(x='horsepower',data=df)
plt.axvline(115,c='r')
plt.axhline(0.5,c='r')
```

```python
fig=plt.gcf()
fig.set_size_inches(12,8)
ax=sns.ecdfplot(x='horsepower',data=df,stat='count')
```
