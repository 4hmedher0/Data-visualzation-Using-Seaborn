```python
import pandas as pd 
import numpy as np 
import seaborn as sns
import matplotlib.pyplot as plt 
```

```python
df=pd.read_csv('Iris.csv')
```

```python
d1=df.iloc[0:4]
d2=df.iloc[55:60]
d3=df.iloc[120:125]
df1=pd.concat([d1,d2,d3])
df1.reset_index(inplace=True)
df1.drop(['index','Id'],axis=1,inplace=True)
df1
```

```python
fig=plt.gcf()
fig.set_size_inches(12,8)
sns.set(style='darkgrid',font_scale=1.3)
ax=sns.catplot(kind='strip',y='PetalWidthCm',x='Species',col='Species',row='Species',data=df1,s=12,linewidth=2,palette='Blues_d',alpha=0.7)
ax.set_titles(col_template='{col_name} Species')
```

```python
cars = sns.load_dataset('mpg').dropna()
cars['type']=['old' if x <=76 else 'new' for x in cars.model_year]
cars['heavy']=['fat' if x >=3000   else('sport' if x<2000  else 'fit') for x in cars.weight]
cars.groupby('heavy').count()
a=sns.countplot(x='heavy',data=cars)
```

```python
ax2=sns.catplot(kind='violin',x='cylinders',y='horsepower',data=cars,col='origin',row='heavy',palette='summer')
```
