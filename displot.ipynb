```python
import pandas as pd 
import numpy as np 
import seaborn as sns
import matplotlib.pyplot as plt 
```

```python
df=sns.load_dataset('mpg').dropna()
```

```python
df['type']=['heavy' if x>2500 else 'fit' for x in df['weight'] ]
df
```

```python
fig=plt.gcf()
fig.set_size_inches(17,20)
ax=sns.displot(x='horsepower',data=df,kde=True,rug=True,lw=2,edgecolor='r')
```
