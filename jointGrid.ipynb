```python
import numpy as np 
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=sns.load_dataset('mpg').dropna()
```

```python
sns.set(style='darkgrid')
df.head(5)
```

```python
fig=plt.gcf()
fig.set_size_inches(12,8)
g=sns.JointGrid(data=df,x='horsepower',y='mpg') 
g.plot(sns.scatterplot,sns.histplot)

```

```python
g=sns.JointGrid(data=df,x='horsepower',y='mpg') 
g.plot(sns.lineplot,sns.histplot)
```

```python
g=sns.JointGrid(data=df,x='horsepower',y='mpg') 
g.plot_joint(sns.histplot)
```

```python
g=sns.JointGrid(data=df,x='horsepower',y='mpg') 
g.plot_marginals(sns.histplot)
```

```python
g=sns.JointGrid(data=df,x='horsepower',y='mpg') 
g.plot_marginals(sns.histplot,element='poly',fill=False)
```

```python
g=sns.JointGrid(data=df,x='horsepower',y='mpg') 
g.plot_marginals(sns.histplot,element='poly',fill=False,lw=3,color='g')
g.plot_joint(sns.scatterplot,color='c')
```

```python
g=sns.JointGrid(data=df,x='horsepower',y='mpg') 
g.plot_marginals(sns.histplot,element='poly',fill=False,lw=3,color='g')
g.plot_joint(sns.kdeplot,color='c')
```

```python
x='horsepower'
y='mpg'
g=sns.JointGrid(data=df,x='horsepower',y='mpg') 
sns.regplot(x=x,y=y,data=df,ax=g.ax_marg_x)
```

```python
x='horsepower'
y='mpg'
g=sns.JointGrid(data=df,x='horsepower',y='mpg',height=9) 
sns.scatterplot(x=x,y=y,data=df,ax=g.ax_marg_y)
```

```python
x='horsepower'
y='mpg'
g=sns.JointGrid(data=df,x='horsepower',y='mpg',height=9) 
sns.kdeplot(x=x,y=y,data=df,ax=g.ax_joint)
```

```python
x='horsepower'
y='mpg'
g=sns.JointGrid(data=df,x='horsepower',y='mpg',height=9) 
sns.kdeplot(x=x,y=y,data=df,ax=g.ax_joint)
sns.histplot(x=x,y=y,data=df,ax=g.ax_marg_x)
sns.regplot(x=x,y=y,data=df,ax=g.ax_marg_y)
```

```python
g=sns.JointGrid(data=df,x='horsepower',y='mpg',hue='origin',height=9) 
g.plot_marginals(sns.histplot,element='poly',fill=False,lw=2,color='g')
g.plot_joint(sns.histplot,color='c')
```

```python
x='horsepower'
y='mpg'
g=sns.JointGrid(data=df,x='horsepower',y='mpg',palette='rocket',height=9) 
sns.histplot(x=x,y=y,data=df,ax=g.ax_joint)
sns.kdeplot(x=x,y=y,data=df,ax=g.ax_marg_x)
sns.regplot(x=x,y=y,data=df,ax=g.ax_marg_y)
```

```python
x='horsepower'
y='mpg'
g=sns.JointGrid(data=df,x='horsepower',y='mpg',palette='rocket',height=9) 
sns.histplot(x=x,y=y,data=df,ax=g.ax_joint)
sns.kdeplot(x=x,y=y,data=df,ax=g.ax_marg_x)
sns.regplot(x=x,y=y,data=df,ax=g.ax_marg_y)
g.set_axis_labels(xlabel='HORSEPOWER',ylabel='MILEPERGALLON')
g.savefig('joingfig.png')
```

```python
g=sns.JointGrid(data=df,x='horsepower',y='mpg',hue='origin',height=9) 
g.plot_marginals(sns.histplot,element='poly',fill=True,lw=1,color='g')
g.plot_joint(sns.histplot,color='c')
```

```python
g=sns.JointGrid(data=df,x='horsepower',y='mpg',height=9, ratio=5, marginal_ticks=True) 
g.plot_marginals(sns.histplot,kde=True)
```

```python
g=sns.JointGrid(data=df,x='horsepower',y='mpg',hue='origin',height=9,marginal_ticks=True) 
g.plot_marginals(sns.histplot,element='poly',fill=False,lw=2,color='g')
g.plot_joint(sns.scatterplot)
g.set_axis_labels(xlabel='xx',ylabel='yy')
g.savefig ('joint.png')
```
