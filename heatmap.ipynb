```python
import pandas as pd 
import numpy as np 
import seaborn as sns
import matplotlib.pyplot as plt 
```

```python
df=sns.load_dataset('mpg').dropna()
```

```python
df.head()
```

```python
df1=df.groupby('origin').cylinders.value_counts().unstack()
```

```python
df1.fillna(0,inplace=True)
```

```python
df1=df1.T
```

```python
df1
```

```python
fig=plt.gcf()
fig.set_size_inches(12,8)
sns.set(style='dark',font_scale=1.4)
ax=sns.heatmap(df1,cmap='Reds',lw=1,vmin=20,vmax=80)
```

```python
fig=plt.gcf()
fig.set_size_inches(12,8)
sns.set(style='dark',font_scale=1.4)
ax=sns.heatmap(df1,cmap='Reds',lw=1,vmin=20,vmax=80,center=90)
```

```python
fig=plt.gcf()
fig.set_size_inches(12,8)
sns.set(style='dark',font_scale=1.4)
ax=sns.heatmap(df1,cmap='Reds')
```

```python
fig=plt.gcf()
fig.set_size_inches(12,8)
sns.set(style='dark',font_scale=1.4)
ax=sns.heatmap(df1,cmap='Reds',lw=1,annot=True)
```

```python
fig=plt.gcf()
fig.set_size_inches(12,8)
sns.set(style='dark',font_scale=1.4)
ax=sns.heatmap(df1,cmap='Reds',lw=1,annot=True,fmt='.0f')
```

```python
fig=plt.gcf()
fig.set_size_inches(12,8)
sns.set(style='dark',font_scale=1.4)
ax=sns.heatmap(df1,cmap='Reds',lw=1,linecolor='g',annot=True,fmt='.0f') 
```

```python
fig=plt.gcf()
fig.set_size_inches(12,8)
sns.set(style='dark',font_scale=1.4)
ax=sns.heatmap(df.corr(),cmap='Reds',lw=1,linecolor='w',annot=True,fmt='.0f') 
```
